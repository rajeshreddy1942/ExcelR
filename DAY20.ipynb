```python
import nltk
import spacy
from nltk.corpus import stopwords

# Download stopwords if not already downloaded
nltk.download('stopwords')

def process_text(text):
    # Load spaCy's English model
    nlp = spacy.load('en_core_web_sm')
    
    # Convert text to lowercase using spaCy
    doc = nlp(text.lower())
    
    # Tokenize and remove stopwords using NLTK
    stop_words = set(stopwords.words('english'))
    filtered_words = [token.text for token in doc if token.text not in stop_words and token.is_alpha]
    
    return filtered_words

# Example usage
text = input("Enter a text: ")
processed_text = process_text(text)
print("Processed words:", processed_text)
```
